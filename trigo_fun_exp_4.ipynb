```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y=np.cosh(x)

plt.plot(x,y,label='cosh')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

figure,axis = plt.subplots(2,2)

x=np.linspace(0,10,100)

y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.sinh(x)
axis[0,0].plot(x,y1)
axis[0,1].plot(x,y2)
axis[1,0].plot(x,y3)
axis[1,1].plot(x,y4)

plt.show()
```
